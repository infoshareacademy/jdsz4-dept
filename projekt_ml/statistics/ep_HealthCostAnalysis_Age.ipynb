```python
x = np.array(df['age'])
y = np.array(df['charges'])

model = LinearRegression()
x = x.reshape(-1,1)
y = y.reshape(-1,1)
model.fit(x,y)
model.score(x,y)

a = model.coef_[0]
b = model.intercept_[0]

def linearFunction(a,b,x):
    return a*x+b

x_reg = x
y_reg = linearFunction(a,b,x)

fig = plt.figure(figsize=(20,20))
plt.plot(x,y,linestyle='none', marker='^', markersize=8, label='age')
plt.plot(x_reg, y_reg, label='Regresion Model')
plt.xlabel('age')
plt.ylabel('charges')
plt.legend(loc='upper left')

print("a: ", a, "  ", "b: ", b)
print("model score: ", model.score(x,y))
print("r^2: ", r2_score(y,y_reg))
```
