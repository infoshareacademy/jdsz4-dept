```python
import pandas as pd
data_frame = pd.read_csv("/home/pawcio/Desktop/big_data_science_course/Projekty/tabele_zawody/WA_Fn-UseC_-HR-Employee-Attrition.csv")
total_base = (data_frame[['MonthlyIncome', 'EducationField', 'Education', 'TotalWorkingYears', 'Department', 'JobRole']])
print(total_base)
```

```python
def applikacja():
    print('Witaj! Właśnie tu możesz poznać swoją przyszłość!\nZapraszamy!\n')
    def get_income_limits(user_monthlyIncome):
        limits = [
            (1000, 3000),
            (3000, 5000),
            (5000, 7000),
            (7000, 9000),
            (9000, 11000),
            (11000, 13000),
            (13000, 15000),
            (15000,20001)
        ]
        for min_val, max_val in limits:
            if min_val < user_monthlyIncome <= max_val:
                return min_val, max_val

    while True:
        try:
            user_monthlyIncome = int(input('Jakie są Twoje wymarzone zarobki? $$$: '))
            print((' '))
            if 1000 <= user_monthlyIncome < 20001:
                break
        except:
            print('Bez żartów, podaj liczbę!')
        else:
            print('Wybierz stawkę z zakresu 1000-20000!')

    monthlyincome_min_limit, monthlyincome_max_limit = get_income_limits(user_monthlyIncome)
    #print(monthlyincome_min_limit, monthlyincome_max_limit)
    filtr_1 = (total_base['MonthlyIncome'] > monthlyincome_min_limit) & (total_base['MonthlyIncome'] < monthlyincome_max_limit)
    df_1_filtr = total_base[filtr_1]

    def get_years_limits(user_totalworkingyears):
        limits = [
            (0, 2),
            (0, 3),
            (0, 4),
            (0, 6),
            (0, 11),
            (0, 16),
            (0, 21),
            (0, 31),
            (0, 41)   
        ]
        for min_value, max_value in limits:
            if min_value <= user_totalworkingyears <= max_value:
                return min_value, max_value

    while True:
        try:
            user_totalworkingyears = int(input('Jaki jest Twój staż pracy? - podaj w liczbach całkowitych: '))
            print((' '))
            if 0 <= user_totalworkingyears < 40:
                break
        except:
            print('Podaj liczbę całkowitą!')
    totalyears_min_limit, totalyears_max_limit = get_years_limits(user_totalworkingyears)
#print(totalyears_min_limit, totalyears_max_limit)
    filtr_2 = (df_1_filtr['TotalWorkingYears'] > totalyears_min_limit) & (df_1_filtr['TotalWorkingYears'] < totalyears_max_limit)
    df_2_filtr = df_1_filtr[filtr_2]
#print(df_2_filtr)

    while True:
        print('Jaka jest Twoja edukacja?')
        print('Life Sciences')
        print('Marketing')
        print('Technical Degree')
        print('Human Resources')
        print('Medical')
        print('Other')
        user_educationfield = input("Wybieram: ")
        print((' '))
        if user_educationfield in ['Life Sciences']:
            break
        elif user_educationfield in ['Marketing']:
            break
        elif user_educationfield in ['Technical Degree']:
            break
        elif user_educationfield in ['Human Resources']:
            break
        elif user_educationfield in ['Medical']:
            break
        elif user_educationfield in ['Other']:
            break
        else:
            print('!!!Skorzystaj z powyższej listy i wybierz odpowiedni obszar!!!')
            print('')
            continue
        
    filtr_3 = df_2_filtr['EducationField'] == str(user_educationfield)
    df_3_filtr = df_2_filtr[filtr_3]
    #print((df_3_filtr))
    while True:
        print('Jaki jest Twój poziom wykształcenia?:')
        print('1. podstawowe')
        print('2. średnie')
        print('3. inżynier/licencjat')
        print('4. magister')
        print('5. doktor')
        user_education = int(input("Wybieram: "))
        if user_education in [1, 2, 3, 4, 5]:
            if user_education == 1:
                break
            if user_education == 2:
                break
            if user_education == 3:
                break
            if user_education == 4:
                break
            if user_education == 5:
                break
            else:
                print('Podaj cyfrę odpowiadającą Twojemu wykształceniu!')
                continue

    filtr_4 = df_3_filtr['Education'] == user_education
    df_4_filtr = df_3_filtr[filtr_4]
    print((' '))
    print('Oto propozycje stanowisk dla Ciebie:')
    #print(df_4_filtr)
    print(list(df_4_filtr['JobRole'].unique()))
    print((' '))
    print('W takim oto Departamencie: ')
    print(list(df_4_filtr['Department'].unique()))
def następna_os():
    while True:
        nastepny = input(('\nCzy jest kolejna osoba, która chce skorzystać z aplikacji? (t/n)\n'))
        if nastepny in ['t', 'T',]:
            applikacja()
        elif nastepny in ['n','N']:
            print('Dziękujemy za skorzystanie, Do widzenia!')
            break
```
